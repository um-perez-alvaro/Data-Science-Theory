```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
b = 1/10
```

```python
def f(x,y):
    return x**2 + b*y**2
def grad_f(x,y):
    return np.array([2*x,2*b*y])
```

```python
x = np.linspace(-2*b,2*b,100)
y = np.linspace(-1.5,1.5,100)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
plt.contour(X,Y,Z,levels=50)
plt.scatter(b,1,color='blue')
```

```python
plt.contour(X,Y,Z,levels=50)
plt.scatter(b,1,color='blue')

#learning rate
learning_rate = 0.9

# starting point
x = np.array([b,1]) 
plt.scatter(x[0],x[1],color='blue')

# Gradient Descent
epochs = 30 #number of GD iterations
for epoch in range(epochs): 
    gradient = grad_f(x[0],x[1])
    new_x = x-learning_rate*gradient # gradient descent update
    plt.scatter(new_x[0],new_x[1],color='blue')
    plt.plot([x[0],new_x[0]],[x[1],new_x[1]],'r--')
    x = new_x
```

```python
k = np.arange(0,50)#.reshape(-1,1)
xs = b*(1-2*learning_rate)**k
ys = (1-2*learning_rate*b)**k
```

```python
plt.contour(X,Y,Z,levels=50)
plt.scatter(b,1,color='blue')

#learning rate
learning_rate = 0.9

# starting point
x = np.array([b,1]) 
plt.scatter(x[0],x[1],color='blue')

# momentum parameter
beta = 0

# initialize directions
d_current = np.zeros(2)
d_previous = np.zeros(2)

# gradient descent with momentum
epochs = 30
for epoch in range(epochs): 
    
    d_current = grad_f(x[0],x[1]) + beta*d_previous
    new_x = x- learning_rate*d_current
    
    plt.scatter(new_x[0],new_x[1],color='blue')
    plt.plot([x[0],new_x[0]],[x[1],new_x[1]],'r--')
    
    x = new_x
    d_previous = d_current
```
