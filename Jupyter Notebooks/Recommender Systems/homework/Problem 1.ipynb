# Problem 1: Book Recommendations

The [Book-Crossings dataset](http://www2.informatik.uni-freiburg.de/~cziegler/BX/) is a book ratings dataset compiled by Cai-Nicolas Ziegler from the [Book-Crossing community](https://www.bookcrossing.com). 
It contains 1.1 million ratings of 270,000 books by 90,000 users. 
The ratings are on a scale from 1 to 10.

Click [here](http://www2.informatik.uni-freiburg.de/~cziegler/BX/BX-CSV-Dump.zip) to download the dataset.
The data consists of three files: ratings (BX-Book-Ratings.csv), books info (BX-Books.csv), and users info (BX-Users.csv)

**Ratings data**

The ratings data set provides a list of ratings that users have given to books. It includes 1,149,780 records and 3 fields: userID, ISBN, and bookRating.

```python
# load the data (you may need to change the path to the file)
path = 'BX-Book-Ratings.csv'
ratings = pd.read_csv(path, sep=';',error_bad_lines=False, encoding="latin-1")
ratings
```

**Books data**

The books data set provides book details. It includes 271,360 records and 8 fields: ISBN, book title, book author, publisher and so on.

```python
path = 'BX-Books.csv'
books = pd.read_csv(path, sep=';', error_bad_lines=False, encoding="latin-1")
books
```

**Users data**

This dataset provides the user demographic information.
It includes 278,858 records and 3 fields: user id, location and age.

```python
path = 'BX-Users.csv'
users = pd.read_csv(path, sep=';',  error_bad_lines=False, encoding="latin-1")
users
```

## Part 1

Find the number of ratings each book received (pandas built-in function [value_counts](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.value_counts.html) may be useful here)

```python
counts = ratings['ISBN'].value_counts()
counts
```

## Part 2

To ensure statistical significance, drop books with less than 100 ratings (pandas built-in functions [index](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.index.html) and [isin](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.isin.html) may be useful here)

```python
ratings = ratings[ratings['ISBN'].isin(counts[counts>=100].index)]
ratings
```
