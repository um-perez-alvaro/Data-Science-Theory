```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('ratings_large.csv')
data
```

```python
# rating matrix
rating_matrix = data.pivot(index='title',columns='userId',values='rating')
rating_matrix
```

```python
# select a user
user_ratings = rating_matrix[4]
user_ratings # we'll predict the missing ratings
```

```python
# or use your fake user (if you created one)
url = 'https://raw.githubusercontent.com/um-perez-alvaro/Data-Science-Theory/master/Data/fake_user.csv'
user_ratings = pd.read_csv(url,index_col='title').squeeze(True)
user_ratings
```

```python
# drop the user from the rating matrix
rating_matrix = rating_matrix.drop(4,axis=1)
```

```python
# movies rated by the user
list(user_ratings[user_ratings.notna()].index)
```

```python
# movies not rated by the user
list(user_ratings[user_ratings.isna()].index)
```

```python
# top 20 movies 
user_ratings.sort_values(ascending=False).head(20)
```

```python
# user z scores
user_mean = user_ratings.mean()
user_std = user_ratings.std()
user_zscores = (user_ratings-user_mean)/user_std
```

```python
user_mean, user_std
```

```python
# z scores
zscores_matrix = (rating_matrix-rating_matrix.mean())/rating_matrix.std()
```

```python
# similarity between "our user" and the other users (we'll use correlation similarity)
similarities = zscores_matrix.corrwith(user_zscores)
```

```python
# select an unrated movie
item = 'World War Z (2013)'
```

```python
# top k similarities (only among the users that have rated item)
k = 10
knn_sim = similarities[zscores_matrix.loc[item].notna()].sort_values(ascending=False).head(k)
knn_sim
```

```python
knn_users = knn_sim.index
knn_users
```

```python
knn_zscores = zscores_matrix.loc[item,knn_users]
total = np.sum(np.abs(knn_sim))
```

```python
# prediction
prediction = user_mean + user_std*(np.sum(knn_zscores*knn_sim)/total)
prediction
```

```python
# predictions for all the unrated movies
user_ratings_pred = pd.Series(index=user_ratings[user_ratings.isna()].index)
user_ratings_pred
```

```python
for item in user_ratings_pred.index:
    # knn similarities
    knn_sim = similarities[zscores_matrix.loc[item].notna()].sort_values(ascending=False).head(k)
    # knn users
    knn_users = knn_sim.index
    # knn zscores
    knn_zscores = zscores_matrix.loc[item,knn_users]
    # denominator
    total = np.sum(np.abs(knn_sim))
    # prediction
    user_ratings_pred[item] = user_mean + user_std*(np.sum(knn_zscores*knn_sim)/total)
```

```python
# top 20 recommendations
user_ratings_pred.sort_values(ascending=False).head(20)
```

```python
# bottom 20 recommendations
user_ratings_pred.sort_values(ascending=False).tail(20)
```

## How good are our predictions?

```python
user_ratings = user_ratings.to_frame()
user_ratings['prediction'] = np.NaN
user_ratings
```

```python
for item in user_ratings.index:
    # knn similarities
    knn_sim = similarities[zscores_matrix.loc[item].notna()].sort_values(ascending=False).head(k)
    # knn users
    knn_users = knn_sim.index
    # knn zscores
    knn_zscores = zscores_matrix.loc[item,knn_users]
    # denominator
    total = np.sum(np.abs(knn_sim))
    # prediction
    user_ratings.loc[item,'prediction'] = user_mean + user_std*(np.sum(knn_zscores*knn_sim)/total)
```

```python
user_ratings['error'] = user_ratings['rating']-user_ratings['prediction']
user_ratings[user_ratings.rating.notna()]
```

```python
# histogram
user_ratings.error.hist()
```

```python
# (square root of the) mean squared error
num_rated_movies = np.sum(user_ratings.rating.notna())
np.sqrt(np.sum(user_ratings.error**2)/num_rated_movies)
```

## Penalizing well-known movies

```python
num_ratings = data.groupby('title').rating.count()
```

```python
num_ratings.plot.hist(bins=20)
```

```python
# popular movies
num_ratings[num_ratings>20000]
```

```python
# weights
num_users = data.userId.nunique()
weights = np.log(num_users/num_ratings)
weights
```

```python
weights['Matrix, The (1999)']
```

```python
weights['Dark City (1998)']
```

```python
# pandas does not have a built-in function that computes weighted correlations
def weighted_corrwith(y):
    # computes weighted correlation between x and y
    x = user_zscores
    
    # mean values
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    
    # numerator
    num = np.sum(weights*(x-x_mean)*(y-y_mean))
    # denominator
    den1 = np.sqrt(np.sum(weights*(x-x_mean)**2))
    den2 = np.sqrt(np.sum(weights*(y-y_mean)**2))
    den = den1*den2
    
    return num/den
```

```python
# similarities
similarities = zscores_matrix.apply(weighted_corrwith)
```

```python
user_ratings['prediction 2'] = np.NaN
for item in user_ratings.index:
    # knn similarities
    knn_sim = similarities[zscores_matrix.loc[item].notna()].sort_values(ascending=False).head(k)
    # knn users
    knn_users = knn_sim.index
    # knn zscores
    knn_zscores = zscores_matrix.loc[item,knn_users]
    # denominator
    total = np.sum(np.abs(knn_sim))
    # prediction
    user_ratings.loc[item,'prediction 2'] = user_mean + user_std*(np.sum(knn_zscores*knn_sim)/total)
```

```python
# top 20 recommendations
user_ratings[user_ratings.rating.isna()]['prediction 2'].sort_values(ascending=False).head(20)
```
