## 1. Latent Factor Model

```python
def latent_factor_model(R, k, learning_rate, n_epochs):
    m, n = R.shape
    
    # number of nonzero ratings
    n_zero_ratings = np.sum(~np.isnan(R))
    
    # initialization
    U = np.random.randn(m,k)
    V = np.random.randn(n,k)
    
    mean_error = np.zeros(n_epochs)
    
    # gradient descent steps
    for i in range(n_epochs):
        
        # error matrix
        E = R-U.dot(V.T)
        E[np.isnan(E)]=0

        # update U and V
        new_U = U + learning_rate*E.dot(V)
        new_V = V + learning_rate*E.T.dot(U)

        U = new_U
        V = new_V
        
        # compute mean_error
        error_squared = np.sum(E**2)
        mean_error[i] = np.sqrt(error_squared/(n_zero_ratings))
        
    return U, V, mean_error
```

**Simple example**

```python
R = np.array([
    [2, np.nan, 3, np.nan],
    [np.nan, 5, np.nan, 4],
    [1, 3, np.nan, 5],
    [np.nan, 4, 2, np.nan]
])
R
```

```python
U, V, mean_error = latent_factor_model(R, 
                                       k = 2,
                                       learning_rate = 0.05,
                                       n_epochs = 50)
```

```python
plt.plot(mean_error)
```

```python
R
```

```python
# approximation of R
U.dot(V.T)
```

## 2. Incorporating user and item biases

```python
def latent_factor_model(R, k, learning_rate, n_epochs):
    m, n = R.shape
    
    # number of nonzero ratings
    n_zero_ratings = np.sum(~np.isnan(R))
    
    # initialization
    U = np.random.randn(m,k+2)
    U[:,k+1] = 1 # set (k+2)th column to 1
    V = np.random.randn(n,k+2)
    V[:,k] = 1 # set (k+1)th column to 1
    
    mean_error = np.zeros(n_epochs)
    
    # gradient descent steps
    for i in range(n_epochs):
        
        # error matrix
        E = R-U.dot(V.T)
        E[np.isnan(E)]=0

        # update U and V
        new_U = U + learning_rate*E.dot(V)
        new_U[:,k+1] = 1
        new_V = V + learning_rate*E.T.dot(U)
        new_V[:,k] = 1

        U = new_U
        V = new_V
        
        # compute mean_error
        error_squared = np.sum(E**2)
        mean_error[i] = np.sqrt(error_squared/(n_zero_ratings))
        
    return U, V, mean_error
```

```python
U, V, mean_error = latent_factor_model(R, 
                                       k = 10, 
                                       learning_rate = .00001, 
                                       n_epochs = 50)
```
