```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from skimage import color, feature
from skimage import data, transform
```

```python
from scipy import signal
```

```python
image = data.coffee()[:,:,0]
image.shape
```

```python
plt.imshow(image, cmap='gray')
```

```python
kernel = np.array([[-1,-1,-1],
                   [0,0,0,],
                   [1,1,1]])
kernel
```

```python
image.shape
```

```python
output = signal.correlate2d(image,kernel,'valid')
plt.imshow(output, cmap='gray')
```

```python
kernel = kernel.T
kernel
```

```python
output = signal.correlate2d(image,kernel,'valid')
plt.imshow(output, cmap='gray')
```
