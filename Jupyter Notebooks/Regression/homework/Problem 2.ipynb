# Problem 2: Contour plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

A **contour** plot (sometimes called a level plot) is a graphical technique for representing a 3-dimensional function $z=f(x,y)$ by plotting constant $z$ slices, called contours, on a 2-dimensional format.
That is, given a value for $z$, lines are drawn for connecting the $(x,y)$ coordinates where that $z$ value occurs.

The basic syntax for creating contour plots is ${\tt plt.contour(X,Y,Z,levels)}$.
To trace a contour, ${\tt plt.contour}$ requires a 2-D array Z that specifies function values on a grid. 
The underlying grid is given by X and Y.
The ${\tt numpy.meshgrid}$ function is useful for constructing 2-D grids from two 1-D arrays. 
It returns two 2-D arrays X,Y of the same shape, where each element-wise pair specifies an underlying (x, y) point on the grid. Function values on the grid Z can then be calculated using these X,Y element-wise pair.

## Example

Let's consider the function 

$$
z = f(x,y) = \sqrt{x^2+y^2}
$$

```python
xlist = np.linspace(-10.0, 10.0, 500) # Create 1-D arrays for x,y axes
ylist = np.linspace(-10.0, 10.0, 500)
X,Y = np.meshgrid(xlist, ylist) # Create 2-D grid xlist,ylist values
Z = np.sqrt(X**2 + Y**2) # Compute function values on the grid
```

We need to specify the contour levels (of Z) to plot. 
You can either specify a positive integer for the number of automatically decided contours to plot, or you can give a list of contour (function) values in the levels argument.

```python
plt.contour(X, Y, Z, levels = 20) # draw 20 contours
```

You can add a colorbar to the plot

```python
plt.contour(X, Y, Z, levels = 20) # draw 20 contours
plt.colorbar()
```

You can draw filled contours

```python
plt.contourf(X, Y, Z, levels = 20) # draw 20 filled contours
plt.colorbar()
```

You can use a different color map ([list of colormaps](https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html))

```python
plt.contourf(X, Y, Z, levels = 20, cmap = 'RdGy') # draw 20 filled contours
plt.colorbar() # add a colorbar
```
