## Part 1: two-dimensional plots

In this exercise, you will learn how to plot a function $y=f(x)$, over a specified interval.

**Example**: Let's plot the **sine function** from âˆ’3 to 2. 
To do so, we need to provide a discretization (grid) of the values along the x-axis, and evaluate the function on
each x value. 

```python
x_plot = np.linspace(-3, 2, 100) # 100 evenly spaced samples, calculated over the interval [-3, 2].
y_plot = np.sin(x_plot) # Evaluate function on x_plot
plt.plot(x_plot, y_plot) # Create line plot with y_plot against x_plot
```

[Here](https://numpy.org/doc/stable/reference/routines.math.html) and [here](https://numpy.org/doc/stable/reference/constants.html) are the lists of all mathematical functions ($\sin$, $\cos$, $\exp$, etc) and constants ($\pi$, e, etc) available in Numpy.

**Remark:** Numpy's mathematical functions are **vectorized functions**.
This means that if you pass a vector

$$
v = 
\begin{bmatrix}
    v_1\\v_2\\ \vdots \\ v_m
\end{bmatrix}
$$

to a function $f(x)$, Numpy will evaluate $f(x)$ at each entry of $v$:

$$
f(v) = 
\begin{bmatrix}
    f(v_1) \\f(v_2)\\ \vdots \\ f(v_m)
\end{bmatrix}
$$

```python
# example
v = np.array([0,np.pi/2,np.pi,3*np.pi/2])
np.sin(v)
```

## Part 2: three-dimensional plots

Three-dimensional plots are enabled by importing the mplot3d toolkit. Once this submodule is imported, a three-dimensional axes can be created by passing the keyword `projection='3d'`.

```python
from mpl_toolkits import mplot3d
```

The basic syntax for creating surface plots is `.plot_surface(X,Y,Z)`. A surface plot requires all the input data to be in the form of two-dimensional regular grids, with the Z data evaluated at each point. The ``numpy.meshgrid`` function is useful for constructing 2-D grids from two 1-D arrays. 

**Your turn:** Let's plot the function $z=\sin\left(\sqrt{x^2+y^2}\right)$ with $x\in[-6,6]$ and $y\in[-6,6]$

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes(projection='3d')


m_plot = 100
x_plot = np.linspace(-6,6,m_plot) 
y_plot = np.linspace(-6,6,m_plot)

X_plot,Y_plot = np.meshgrid(x_plot,y_plot)

Z_plot = np.sin(np.sqrt(X_plot** 2 + Y_plot** 2))

ax.plot_surface(X_plot,Y_plot,Z_plot)
```

Adding a colormap to the surface plot can aid perception of the topology of the surface being visualized:

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes(projection='3d')


m_plot = 100
x_plot = np.linspace(-6,6,m_plot) 
y_plot = np.linspace(-6,6,m_plot)

X_plot,Y_plot = np.meshgrid(x_plot,y_plot)

Z_plot = np.sin(np.sqrt(X_plot** 2 + Y_plot** 2))

ax.plot_surface(X_plot,Y_plot,Z_plot,cmap='viridis')
```

## Part 3: contour plots

A **contour** plot (sometimes called a level plot) is a graphical technique for representing a 3-dimensional function $z=f(x,y)$ by plotting constant $z$ slices, called contours, on a 2-dimensional format.

The basic syntax for creating contour plots is `.contour(X,Y,Z,levels)`. A contour plot requires all the input data to be in the form of two-dimensional regular grids, with the Z data evaluated at each point.

**Example:** Let's consider the function $z=\sin\left(\sqrt{x^2+y^2}\right)$ with $x\in[-6,6]$ and $y\in[-6,6]$

```python
m_plot = 100
x_plot = np.linspace(-6,6,m_plot) 
y_plot = np.linspace(-6,6,m_plot)

X_plot,Y_plot = np.meshgrid(x_plot,y_plot)

Z_plot = np.sin(np.sqrt(X_plot** 2 + Y_plot** 2))
```

We need to specify the contour levels (of Z) to plot. You can either specify a positive integer for the number of automatically decided contours to plot, or you can give a list of contour (function) values in the levels argument.

```python
plt.contour(X_plot, Y_plot, Z_plot, levels = 20) # draw 20 contours
```

You can add a colorbar to the plot

```python
plt.contour(X_plot, Y_plot, Z_plot, levels = 20) # draw 20 contours
plt.colorbar()
```

You can draw filled contours

```python
plt.contourf(X_plot, Y_plot, Z_plot, levels = 20) # draw 20 filled contours
plt.colorbar()
```

You can use a different color map ([list of colormaps](https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html))

```python
plt.contourf(X_plot, Y_plot, Z_plot, levels = 20, cmap='seismic') # draw 20 filled contours
plt.colorbar()
```
